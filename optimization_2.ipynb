```python
#convert the categorical data in numeric using get_dummies()
dummies = pd.get_dummies(application_df)
# Convert boolean to integer
dummies = dummies.astype(int)

dummies.head(5)
```

```python
# Split our preprocessed data into our features and target arrays

y = dummies['IS_SUCCESSFUL'].values
X = dummies.drop(columns='IS_SUCCESSFUL').values



# Split the preprocessed data into a training and testing dataset

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

compile ,train,evaluate the model


```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer(Third Attempt)
#  YOUR CODE GOES HERE

nn_model = tf.keras.models.Sequential()

# Input features total
input_features_total = len(X_train[0])
hidden_nodes_layer1 = 80
hidden_nodes_layer2 = 30


# First hidden layer
nn_model.add(tf.keras.layers.Dense(units=80, activation='relu', input_shape=(input_features_total,)))

# Second hidden layer
nn_model.add(tf.keras.layers.Dense(units=30, activation='relu'))

# Output layer

nn_model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
# Check the structure of the model
nn_model.summary()
```

```python
# Compile the model (Third Attempt)
nn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model (First Attempt)
nn_model.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data (First Attempt)
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
